```python
brightness_x = brightness_x/255
brightness_ds = td.TensorDataset(torch.Tensor(brightness_x), torch.Tensor(brightness_y))
brightness_dl = td.DataLoader(brightness_ds, batch_size=128) # create your dataloader
```

```python
xc_sample = None
for i, (X, y) in enumerate(brightness_dl):
    xc_sample = X
    if i > 5:
        break
    print(i)
    print(X.shape)
    print(y.shape)
    print(X)
    print(y)
    imshow(xc_sample[0].reshape(32, 32, -1))
    colorbar()
    show()
```
